```python
import matplotlib.pyplot as plt

#製作figure元件(畫布)
fig = plt.figure()

#在figure上製作一個axes(圖表)
ax_1 = fig.add_subplot(1,1,1)

#將圖的數據儲存在axes中
X = [0,1,2] #X軸
Y = [3,4,5] #Y軸
ax_1.plot(X,Y)

#顯示圖
plt.show()
```

```python
import matplotlib.pyplot as plt

figure = plt.figure()
axes = figure.add_subplot(1,1,1) #一列一欄第一張圖表
X = [0,1,2] #X軸
Y = [3,4,5] #Y軸
axes.plot(X,Y,color='pink',marker='o',linestyle='dashed',linewidth='5',markersize='15')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")

student1 = {'國文':78,'英文':80,'數學':92,'歷史':75,'探索':85}
student2 = {'國文':90,'英文':75,'數學':80,'歷史':75,'探索':65}
subjects = list(student1.keys())
stu1_values = list(student1.values())
stu2_values = list(student2.values())
```

```python
#標準寫法
figure = plt.figure(figsize=(8,5),dpi=240,facecolor='#cccccc')
axes = figure.add_subplot(1,1,1)
axes.set_title("學生分數")
axes.set_xlabel("科目")
axes.set_ylabel("分數")
axes.plot(subjects,stu1_values,marker='*',label="學生1",linestyle='-.')
axes.plot(subjects,stu2_values,marker='*',label="學生2",linestyle='-.')
axes.hlines(80,xmin=-1,xmax=6) #水平線


x_locations = axes.get_xticks() #取得X的位置(0~4)
y_locations = axes.get_yticks()
#print(axes.get_xticklabels()) 
#axes.set_xticklabels(['chinese','english','math','history','discover']) #改X欄名
#axes.set_xlim(xmin=-1,xmax=5) #設定X顯示範圍
axes.set_xlim(xmin=min(x_locations)-1,xmax=max(x_locations)+1)
axes.set_ylim(ymin=min(y_locations)-10,ymax=max(y_locations)+10)

#axes.annotate("註解",(0,80))
for x in x_locations:
    axes.annotate(str(stu1_values[x]),(x,stu1_values[x]+1))
    axes.annotate(str(stu2_values[x]),(x,stu2_values[x]+1))

axes.legend()
#axes.grid() #網格
axes.grid(axis='y')
plt.show()
plt.close()
```
