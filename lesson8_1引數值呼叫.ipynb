```python
# 如何呼叫function
```

```python
def menu(wine:str, entree:str, dessert:str):
  print(f"前酒:{wine}")
  print(f"主餐:{entree}")
  print(f"甜點:{dessert}")
```

```python
# 引數值呼叫(依照參數順序)
```

```python
menu("白酒","雞排","蛋糕")
```

```python
# 引數名稱的呼叫(可不照順序)
```

```python
menu(wine="白酒",dessert="蛋糕",entree="雞排")
```

# 混合呼叫
- 前面要引數值呼叫
- 後面要引數名稱的呼叫
- 引數名稱呼叫後，就全部要使用引數名稱

```python
menu("白酒",entree="雞排",dessert="蛋糕")
menu("白酒",dessert="蛋糕",entree="雞排")
```

# 參數有預設值
- 有預設值的參數 一定要放最後面
- 呼叫方式變多


```python
def menu(wine:str, entree:str, dessert:str='奶昔'):
  print(f"前酒:{wine}")
  print(f"主餐:{entree}")
  print(f"甜點:{dessert}")
```

```python
menu("白酒",entree="雞排")
```

```python
menu("白酒",dessert="蛋糕",entree="雞排")
```

```python
#沒有限定數量的引數值呼叫
def print_args(*args):
  print(args)

print_args(1,2,3,'a','V') #tuple型別的引數值
```

```python
def print_args(**kwargs):
  print(kwargs)

print_args(one=1,two=2,theaa=3,A='a') #字典型別的引數值
```

```python
def print_args(*args,**kwargs):
  print(args)
  print(kwargs)

print_args(1,2,3,'a',one=1,A='a')
```

```python
#print(*objects, sep=' ', end='\n', file=None, flush=False)
print(1,2,3,4,sep='-')
print(1,2,3,4,end='.\n',sep='-') #print的預設引數名稱不用照順序
```
