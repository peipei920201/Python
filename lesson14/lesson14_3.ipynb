```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")
```

```python
import random

#list comprehension(for in 簡寫)
days = [day for day in range(1,31)]

#list建構式(初始化)
days = list(range(1,31))

july = [round(random.uniform(28.5,35.2),ndigits=1) for _ in range(1,31)] #生成範圍內的隨機浮點數
august = [round(random.uniform(28.5,35.2),ndigits=1) for _ in range(1,31)]
```

```python
figure = plt.figure(figsize=(8,12))
axes1 = figure.add_subplot(2,1,1)
#線圖plot
axes1.plot(days, july,marker='o',color='brown',linestyle='--',label='7月')
axes1.set_xticks(days)
axes1.grid(axis='y')
axes1.set_title('7月溫度')
axes1.set_xlabel("日期")
axes1.set_ylabel("攝氏")
axes1.legend()

axes2 = figure.add_subplot(2,1,2)
#散佈圖scatter
axes2.scatter(days,august,marker='*',label='八月',color='y')
axes2.set_xticks(days)
axes2.grid(axis='y')
axes2.set_title('8月溫度')
axes2.set_xlabel("日期")
axes2.set_ylabel("攝氏")
axes2.legend()

plt.show()
plt.close()
```
