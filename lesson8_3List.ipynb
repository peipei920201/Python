```python
#組合多筆資料成為有順序的資料(list)
```

```python
#list內的元素 一般都是相同型別
nums:list[int]=[1,2,3,4,5]
nums
```

```python
#取出元素
nums[0]
nums[4]
nums[-1]
```

```python
#切割元素
nums[2:4]
```

```python
nums[:3] #前3個
```

```python
nums[-3:] #後3個
```

```python
#List串列資料 代表可以iterable(重複呼叫)
for i in nums:
  print(i,end=' ')
```
